Ar Vilniaus miegamuosiuose rajonose gyvena daugiau jaunų žmonių nei vyresnių? Pasirinkau jaunu zmonių amžių iki 39 metų imtinai.

```python
import pandas as pd
```

```python
vp = pd.read_csv('registered_people_n_streets.csv')
```

```python
vp
```

```python
vp['amzius'] = (2023 - vp['GIMIMO_METAI'])
vp
```

```python
vp
```

```python
vp['visi'] = 1
vp.loc[vp['amzius'] <= 39, 'jaunimas'] = 1 
vp.loc[vp['amzius'] > 39, 'jaunimas'] = 0 
vp

#
```

```python
vp_sum = vp.groupby('SENIUNIJA').sum()[ [ 'visi', 'jaunimas'] ]
vp_sum
```

```python
vp_sum['vyresni_zmones'] = vp_sum['visi'] - vp_sum['jaunimas']
vp_sum
```

```python
vp_sum['jaunimas_proc'] = vp_sum['jaunimas'] / vp_sum['visi']
vp_sum['vyresni_zmones_proc'] = vp_sum['vyresni_zmones'] / vp_sum['visi']
vp_sum
```

```python
vp_sum.to_csv('gyventojai_vln.csv')
```
